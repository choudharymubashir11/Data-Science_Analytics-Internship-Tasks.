```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

%pip install kagglehub
import kagglehub

# Download the Iris dataset from UCIML via kagglehub [cite: 13]
dataset_path = kagglehub.dataset_download("uciml/iris")
file_path = os.path.join(dataset_path, "Iris.csv")

# Load the dataset using pandas [cite: 14]
df = pd.read_csv(file_path)

# Display dataset structure: shape, columns, and head 
print(f"Dataset Shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")
print("\nFirst 5 rows:")
display(df.head())

# Basic data summarization [cite: 23]
print("\nStatistical Summary:")
display(df.describe())

# Clean the data: Remove 'Id' if it exists [cite: 85]
if 'Id' in df.columns:
    df = df.drop(columns=['Id'])
```

```python
#3. Exploratory Data Analysis (EDA).Using matplotlib and seaborn , we will create the visualizations required by the task.
#Scatter Plot: Relationship Analysis
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='SepalLengthCm', y='SepalWidthCm', hue='Species', palette='viridis')
plt.title('Sepal Length vs Sepal Width')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# Histogram: Data Distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['PetalLengthCm'], kde=True, color='skyblue')
plt.title('Distribution of Petal Length')
plt.xlabel('Petal Length (Cm)')
plt.show()
```

```python
#Box Plot: Outlier Detection
plt.figure(figsize=(10, 6))
sns.boxplot(x='Species', y='PetalWidthCm', data=df)
plt.title('Petal Width Spread and Outliers by Species')
plt.show()
```
